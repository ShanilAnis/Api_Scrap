```python
urls= 'https://www.programmableweb.com/category/all/apis' 
job_no =0
no_jobs ={}

while True:
    
    request = requests.get(urls)
    data = request.text
    soup= BeautifulSoup(data,'html.parser')
    jobs= soup.find_all('tr',{'class': ['odd','even']})
    for job in jobs:
        title = job.find('a').text
        description = job.find('td', {'class':'views-field-field-api-description'}).text
        category = job.find('td', {'class','views-field-field-article-primary-category'}).text
        versions = job.find('td', {'class','views-field-pw-version-links'}).text
        job_no += 1
        no_jobs[job_no] = [title, description, category, versions]
#         print("Titles:", title, "\nDescription", description, "\nCategory", category, "\nVersions", versions, "\n------")
    
    url_tag = soup.find('li',{'class', 'pager-last'}).find('a')
    if url_tag.get('href'):
        urls='https://www.programmableweb.com' + url_tag.get('href')
        print(urls)
    else:
        break

    
  
        
    
    

# title = jobs.find_all('td',{'class','views-field'}).text
# print(title)    
```

```python
api= pd.DataFrame.from_dict(no_jobs, orient='index', columns = ['Title', 'Description', 'Category', 'Versions'])
```
